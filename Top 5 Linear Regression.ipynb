```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
top_five_ethnicity = pd.read_csv('../final-project/data/top_5_HIEF_data.csv')
top_five_ethnicity.set_index("Country")
```

```python
ireland = top_five_ethnicity[(top_five_ethnicity.Country == 'Ireland')]
ireland
```

```python
X = ireland.Year.values.reshape(-1, 1)
y = ireland.EFindex.values.reshape(-1, 1)
```

```python
model = LinearRegression()
model.fit(X, y)
x_min = np.array([[X.min()]])
x_max = np.array([[X.max()]])
print(f"Min X Value: {x_min}")
print(f"Max X Value: {x_max}")
```

```python
y_min = model.predict(x_min)
y_max = model.predict(x_max)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(X, y, c='blue')
plt.plot([x_min[0], x_max[0]], [y_min[0], y_max[0]], c='red')
plt.title("Ireland")
plt.xlabel("Years")
plt.ylabel("EF Index")
```
