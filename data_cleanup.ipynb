```python
# Import dependencies
import pandas as pd
import numpy as np
import glob
```

```python
# Set path (Change to your file path)
path = r'C:\Users\Michael\Documents\GitHub\final-project\data\basic_index'
all_files = glob.glob(path + "/*.csv")

# list to hold df's
li = []

# For loop to create a df for each csv file
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    df = df.fillna(df.mean())
    df = df.round(decimals=2)
    li.append(df)
    
# Concat all csv's into one dataframe
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
# Export to csv
frame.to_csv('data/basic_index/index_concat.csv')
```
